MODEL BUILDING

```python
#For finding the CLV we need lifetimes library
!pip install lifetimes
```

```python
import lifetimes
```

```python
summary=lifetimes.utils.summary_data_from_transaction_data(data_agg,"CustomerID","InvoiceDate","Amount")
```

```python
summary=summary.reset_index()
```

```python
summary.head()
```

**frequency**  
Represents the number of repeat purchases the customer has made.  
This means that its one less than the total number of purchases.  
This is actually slightly misleading  its the count of time periods in which the customer made a purchase.  
For example, if using days as the unit, then it's the number of *distinct days* the customer made a purchase on.

**T**  
Represents the age of the customer in the time units chosen.  
It is equal to the duration between a customers first purchase and the end of the period under study.

**recency**  
Represents the age of the customer when they made their most recent purchase.  
This is equal to the time between a customers first purchase and their most recent purchase.  
If they made only one purchase, the recency is 0.

**monetary_value**  
Represents the average value of a customer's purchases.  
This is calculated as the total monetary value of all the customer's transactions divided by the total number of transactions (not frequency).  
So note that the denominator is different from the frequency described above.


```python
#to predict the number of expected transactions we use BetaGeoFitter
from lifetimes import BetaGeoFitter
bgf=BetaGeoFitter()
bgf.fit(summary['frequency'], summary['recency'], summary['T'])
```

```python
#to find the expected number of transactions for a period "t"
t=30
summary["predicted_num_txn"]=bgf.conditional_expected_number_of_purchases_up_to_time(t,summary["frequency"],summary["recency"],summary["T"])
```

```python
summary.sort_values(by="predicted_num_txn",ascending=False).head()
```

```python
# To use Gamma-Gamma Model for finding CLV, we need to be sure that, there isnt any correlation between frequency and monetary value
summary[["frequency","monetary_value"]].corr()
```

```python
# The monetary_value we obtained is mean, what it means is for the customers with only one transaction or frequency, monetary value will be zero.Thus we need to filter data removing the zero monetay values

summary_filtered=summary[summary["frequency"]>0]
```

```python
from lifetimes import GammaGammaFitter
ggf=GammaGammaFitter()
ggf.fit(summary_filtered["frequency"],summary_filtered["monetary_value"])
```

```python
#to calculate the expexted average profit/revenue per customer(transaction)
summary_filtered["expected_avg_sales"]=ggf.conditional_expected_average_profit(summary_filtered["frequency"],summary_filtered["monetary_value"])
```

```python
summary_filtered.head()
```

```python
summary_filtered['predicted_clv'] = ggf.customer_lifetime_value(
    bgf, summary_filtered['frequency'], summary_filtered['recency'], summary_filtered['T'], summary_filtered['monetary_value'],
    time=1,  # Prediction horizon(months)
    discount_rate=0.01  # Discount rate for future cash flow
)
summary_filtered.head()
```

```python
summary_filtered['manual_predicted_clv'] = summary_filtered['predicted_num_txn'] * summary_filtered['expected_avg_sales']
```

```python
summary_filtered.head()
```

```python
summary_filtered.shape
```

MODEL EVALUATION

```python
from sklearn.metrics import mean_absolute_error, root_mean_squared_error

mae = mean_absolute_error(summary_filtered['predicted_clv'], summary_filtered['manual_predicted_clv'])
rmse = root_mean_squared_error(summary_filtered['predicted_clv'], summary_filtered['manual_predicted_clv'])
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
